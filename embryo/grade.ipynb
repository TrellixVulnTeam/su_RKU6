```python
import os
import glob
from csv import DictReader
import shutil
import glob
import plotnine as p9
import pandas as pd
from pathlib import Path
```

```python
aa = []
ab = []
ac = []
ba = []
bb = []
bc = []
ca = []
cb = []
cc = []
na = []

base_dir = '/data/embryo/day_5/'
with open('clinical_data.csv', 'r') as read_obj:
    csv_dict_reader = DictReader(read_obj)
    for row in csv_dict_reader:
        batch = row['Slide ID']
        batch = batch.replace('.', '_')
        id = row['well.no']
        dir_name = batch + '_pdb_E' + id
        dir_path = Path(base_dir + dir_name)
        if dir_path.exists():
            dir_name = dir_name
        else:
            batch = batch.rsplit('_P', 1)[0]
            dir_name = batch + '_E' + id
            dir_path = Path(base_dir + dir_name)
        grade = row['FreshCellGrade']
        if dir_path.exists():
            if grade == 'aa':
                aa.append(dir_name)
            elif grade == 'ab':
                ab.append(dir_name)
            elif grade == 'ac':
                ac.append(dir_name)
            elif grade == 'ba':
                ba.append(dir_name)
            elif grade == 'bb':
                bb.append(dir_name)
            elif grade == 'bc':
                bc.append(dir_name)
            elif grade == 'ca':
                ca.append(dir_name)
            elif grade == 'cb':
                cb.append(dir_name)
            elif grade == 'cc':
                cc.append(dir_name)
            elif grade == 'NA':
                na.append(dir_name)
        else:
            pass
print('aa: ', len(aa))
print('ab: ', len(ab))
print('ac: ', len(ac))
print('ba: ', len(ba))
print('bb: ', len(bb))
print('bc: ', len(bc))
print('ca: ', len(ca))
print('cb: ', len(cb))
print('cc: ', len(cc))
print('N/A: ', len(na))

```
