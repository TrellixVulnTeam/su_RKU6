```python
import os
import random
import shutil
```

```python
random.seed(8)

train_positive = []
validation_positive = []
train_negative = []
validation_negative = []

method_1_positive = os.listdir('/data/embryo/method_1/positive')
method_1_negative = os.listdir('/data/embryo/method_1/negative')
a = 0
b = 0


while a < 160:
    embryo = random.choice(method_1_positive)
    if embryo not in train_positive:
        train_positive.append(embryo)
        a +=1
for embryo in method_1_positive:
    if embryo not in train_positive:
        validation_positive.append(embryo)
while b < 147:
    embryo = random.choice(method_1_negative)
    if embryo not in train_negative:
        train_negative.append(embryo)
        b +=1
for embryo in method_1_negative:
    if embryo not in train_negative:
        validation_negative.append(embryo)

for image in train_positive:
    shutil.copy('/data/embryo/method_1/positive/' + image, '/data/embryo/method_1/train/positive/' + image)
for image in validation_positive:
    shutil.copy('/data/embryo/method_1/positive/' + image, '/data/embryo/method_1/validation/positive/' + image)
for image in train_negative:
    shutil.copy('/data/embryo/method_1/negative/' + image, '/data/embryo/method_1/train/negative/' + image)
for image in validation_negative:
    shutil.copy('/data/embryo/method_1/negative/' + image, '/data/embryo/method_1/validation/negative/' + image)

    
```
