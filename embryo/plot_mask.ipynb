```python
import plotnine as p9
import pandas as pd
```

```python
data = pd.read_csv("/data/embryo/models/research/model-4-008-636.csv")
```

```python
plot = (p9.ggplot(data = data,
mapping = p9.aes(x = 'Time', y = 'Zona', color = 'ID')))
plot + p9.geom_line() + p9.labs(title = "Zona size over time", x='Time (minutes)', y='Size')
```

```python
plot = (p9.ggplot(data = data,
mapping = p9.aes(x = 'Time', y = 'Zona', color = 'ID')))
plot +p9.geom_smooth(method = 'loess', span = 0.4, se=False) + p9.labs(title = "Zona size over time", x='Time (minutes)', y='Size')
```

```python
data['Embryo_SMA'] = data.iloc[:,4].rolling(window=100).mean()
data['Zona_SMA'] = data.iloc[:,5].rolling(window=3).mean()
data['Ratio_SMA'] = data.iloc[:,6].rolling(window=3).mean()
```

```python
plot = (p9.ggplot(data = data,
mapping = p9.aes(x = 'Time', y = 'Zona_SMA', color = 'ID')))
plot + p9.geom_line() + p9.labs(title = "Zona size over time", x='Time (minutes)', y='Size')
```

```python
x = 0
lst = data['Zona_SMA'].tolist()
for size in lst:
    if size > x:
        x = size
    else:
        pass
index = lst.index(x)
time = data['Time'][index]
id = data['ID'][index]
print(id, index, time, x)
```
