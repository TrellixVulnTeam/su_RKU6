```python
import plotnine as p9
import pandas as pd
```

```python
data = pd.read_csv("/data/embryo/tfrecords/yolov3-tf2/D2018_02_25_S00434_I0776.csv")
zona_data = data[data.Class == 'zona']
embryo_data = data[data.Class == 'embryo']
```

```python
plot = (p9.ggplot(data = zona_data,
mapping = p9.aes(x = 'Time', y = 'Size', color = 'ID')))
plot + p9.geom_smooth(method = 'loess', span = 0.4, se=False)
```

```python
plot = (p9.ggplot(data = embryo_data,
mapping = p9.aes(x = 'Time', y = 'Size', color = 'ID')))
plot + p9.geom_smooth(method = 'loess', span = 0.4, se=False)
```

```python
plot = (p9.ggplot(data = data,
mapping = p9.aes(x = 'Time', y = 'Size', color = 'ID')))
plot + p9.geom_smooth(method = 'loess', span = 0.4, se=False) + p9.facet_wrap('Class')
```
